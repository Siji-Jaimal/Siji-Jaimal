```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
from google.colab.patches import cv2_imshow

```

```python
!pip install scikit-image
```

```python
image =cv2.imread('/content/EcTaNL.jpg', cv2.IMREAD_COLOR)
print("original Noisy image:")
cv2_imshow(image)
```

```python
denoised_image = cv2.GaussianBlur(image, (5,5),0)
print("Denoised Image:")
cv2_imshow(denoised_image)
```

```python
#mean square error
mse = np.mean((image - denoised_image) ** 2)
```

```python
#calculate Structural Similarity Index
ssim_value = ssim(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY),
                  cv2.cvtColor(denoised_image, cv2.COLOR_BGR2GRAY))

print(f"\nMean Squared Error (MSE): {mse}")
print(f"Structural Similarity Index (SSIM:): {ssim_value}")
```
